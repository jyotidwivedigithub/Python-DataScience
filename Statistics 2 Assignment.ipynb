```python
Q1. What are the three measures of central tendency?
Ans.The three measures of central tendency are mean, median, and mode.

Q2. What is the difference between the mean, median, and mode? How are they used to measure the
   central tendency of a dataset?
Ans.The mean, median, and mode are all measures of central tendency, but they focus on different aspects of a dataset.
    1.Mean:-The mean is calculated by adding up all the values in a dataset and then dividing by the total number of values.
      The mean is commonly used when the data is approximately symmetrically distributed and not heavily influenced by outliers.
        
    2.Median:-The median is the middle value in a dataset when it is arranged in ascending or descending order. If there is an 
      even number of observations, the median is the average of the two middle values.The median is useful when dealing with 
        skewed distributions or datasets with outliers.
        
    3.Mode:- The mode is the value that appears most frequently in a dataset. If no value is repeated, the dataset is considered
      to have no mode. If no value is repeated, the dataset is considered to have no mode.
        
Q3. Measure the three measures of central tendency for the given height data:
    [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]
    
Ans. Data: [178, 177, 176, 177, 178.2, 178, 175, 179, 180, 175, 178.9, 176.2, 177, 172.5, 178, 176.5]

    1.Mean:
        Mean = (178+177+176+177+178.2+178+175+179+180+175+178.9+176.2+177+172.5+178+176.5) /16
        Mean = 2832.3/16
        Mean = 177.01875

    2.Median:
        *First, we need to arrange the data in ascending order:
            172.5,175,175,176,176.2,176.5,177,177,177,178,178,178,178.2,178.9,179,180  
        
        *Since there are 16 values, the median is the average of the 8th and 9th values.
            Median = (177+177)/2
            Median = 177
            
    3.Mode:The mode is the value that appears most frequently in the dataset.But in this dataset there is no most frequently
           repeated value.

Q4. Find the standard deviation for the given data:
    [178,177,176,177,178.2,178,175,179,180,175,178.9,176.2,177,172.5,178,176.5]
Ans. 
      Standard Deviation = Square Root of Mean of Squared Differences
    
        Mean = (178+177+176+177+178.2+178+175+179+180+175+178.9+176.2+177+172.5+178+176.5) /16
        Mean = 2832.3/16
        Mean = 177.01875
        
        Mean of Squared Differences= ∑Squared Differences / Number of values
            =[(178-177.1)**2+(177−177.1)**2+(176-177.1)**2+(177−177.1)**2+(178.2−177.1)**2+(178−177.1)**2+(175−177.1)**2+(179−177.1)**2+
             (180−177.1)**2+(175−177.1)**2+(178.9−177.1)**2+(176.2−177.1)**2+(177−177.1)**2+(172.5−177.1)**2+(178−177.1)**2+(176.5−177.1)**2]/16
            =(0.81+0.01+1.21+0.01+1.21+0.81+4.41+3.61+8.41+4.41+3.24+0.81+0.01+21.16+0.81+0.36) / 16
            =3.20

        Standard Deviation = square root of 3.20
                           = 1.71

Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
    the spread of a dataset? Provide an example.
Ans.Measures of dispersion such as range, variance, and standard deviation, are used to describe the spread  of a dataset. They 
    provide insights into how much individual data points deviate from the central tendency measures (mean, median, mode).
    
    1.Range: Range is the difference between the maximum and minimum values in a dataset.
        Example: Consider two datasets:
                Dataset A: [10, 15, 20, 25, 30] (Range = 30 - 10 = 20)
                Dataset B: [5, 15, 25, 35, 45] (Range = 45 - 5 =40)
                Both datasets have the same mean (20), but Dataset B has a larger range, indicating greater dispersion.
                
    2.Variance: Variance is the average of the squared differences between each data point and the mean.
        Example: Using the same datasets A and B:
                Dataset A: Variance = 25 (calculated)
                Dataset B: Variance = 200 (calculated)
                Here, Dataset B has a higher variance, indicating greater variability in the data points.
                
    3.Standard Deviation: Standard deviation is the square root of the variance. It has the same units as the original data.
        Example: Datasets A and B:
                Dataset A: Standard Deviation ≈ 5 (calculated)
                Dataset B: Standard Deviation ≈ 14.14 (calculated)
                The standard deviation emphasizes the greater spread in Dataset B compared to Dataset A.
                
                
Q6. What is a Venn diagram?
Ans. A Venn diagram is a graphical representation used to show the relationships between different sets or groups of items. It consists
    of overlapping circles, each representing a set. The points where the circles overlap indicate elements that are common to the sets, 
    while the non-overlapping parts represent elements that are unique to each set.
    
    
Q7. For the two given sets A = (2,3,4,5,6,7) & B = (0,2,6,8,10). Find:
    (i) A B
    (ii) A ⋃ B
Ans. Set A = {2, 3, 4, 5, 6, 7}
     Set B = {0, 2, 6, 8, 10}

        (1)Intersection (A ∩ B): This represents the set of elements that are common to both Set A and Set B. From the given sets, the common 
        elements are 2 and 6.
             So, A ∩ B = {2, 6}.

        (2)Union (A ∪ B): This represents the set of all elements that are in Set A, Set B, or in both. Combine all elements from both sets without
        repeating any elements.
            So, A ∪ B = {0, 2, 3, 4, 5, 6, 7, 8, 10}.
            
            
Q8. What do you understand about skewness in data?
Ans. Skewness is a statistical concept that describes the asymmetry or lack of symmetry in the distribution of data values within a dataset. It measures
    the extent to which the data is not evenly distributed around the mean, indicating how the tails of the distribution differ from the normal distribution's
    bell-shaped curve. There are three types of skewness...
    (1)Positive Skewness(Right Skewed)
    (2)Negative Skewness (Left-Skewed)
    (3)No Skewness (Symmetrical)
    
    
Q9. If a data is right skewed then what will be the position of median with respect to mean?
Ans. If a dataset is right-skewed (positively skewed), the position of the median will generally be less than the mean. In a right-skewed distribution,
     the tail on the right side (representing the larger values in the dataset) is longer than the tail on the left side. This longer tail pulls the
     mean towards the higher values.
        
        
Q10. Explain the difference between covariance and correlation. How are these measures used in
    statistical analysis?
Ans. Covariance and correlation are both measures that describe the relationship between two variables in statistics, but they differ in terms of scale 
     and interpretability.
    
    (1) Covariance:-Covariance measures the degree to which two variables change together. It indicates whether an increase in one variable corresponds 
        to an increase or decrease in another.
    (2) Correlation:-Correlation is a standardized measure that represents the strength and direction of the linear relationship between two variables. 
        

Q11. What is the formula for calculating the sample mean? Provide an example calculation for a
    dataset.
Ans. The sample mean, denoted as  "X bar" , is the average value of a set of data points in a sample. 
    Suppose we have the following dataset: 10,15,18,22,25
         sample mean = (10+15+18+22+25)/5
                     = 18


Q12. For a normal distribution data what is the relationship between its measure of central tendency?
Ans. For a normal distribution, which is also known as a Gaussian distribution or bell curve, the relationship between its measures of central tendency 
      (mean, median, and mode) is as follows...
    
    (1)Mean (μ):In a normal distribution, the mean is located at the center of the distribution.
                The mean is the balancing point, and the distribution is symmetric around it.
                The mean, median, and mode are all equal in a perfectly normal distribution.
            
    (2)Median:The median is also located at the center of a normal distribution.
              In a symmetrical normal distribution, the median is equal to the mean.
              In any normal distribution, regardless of skewness, the median is close to the mean.
            
    (3)Mode:The mode of a normal distribution is also at the center.
            In a perfectly normal distribution, there is only one mode, and it is equal to the mean.
            For symmetric distributions, the mean and mode coincide.

    For a normal distribution:
                     Mean = Median = Mode
            
            

Q13. How is covariance different from correlation?
Ans. Covariance and correlation are both statistical measures used to describe the relationship between two variables.But both are different from each 
    other....
    1.Covariance: It measures the degree to which two variables vary together. If the variables tend to show similar behavior (both increase or 
      decrease together), the covariance is positive.If one variable tends to increase when the other decreases, the covariance is negative.
      It is scale-dependent; its value is affected by the units of measurement for X and Y. This makes it hard to interpret the strength of the 
        relationship between variables just by looking at the covariance value.
        
    2.Correlation:  It measures both the strength and direction of the linear relationship between two variables. Unlike covariance, correlation is 
      dimensionless and normalized, typically ranging from -1 to +1. The most common measure of correlation is the Pearson correlation coefficient.
      It is dimensionless and standardized. A correlation of +1 indicates a perfect positive linear relationship, -1 indicates a perfect negative linear
      relationship, and 0 indicates no linear relationship, regardless of the units of measurement.
        
        
        
Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.
Ans. Outliers can significantly impact measures of central tendency and dispersion in a dataset. Central tendency measures, such as the mean, are 
     particularly sensitive to outliers, while dispersion measures, like the range or standard deviation, can be influenced by extreme values.
    
     Measures of Central Tendency ( Mean): 
            Example: Consider the dataset [10, 15, 20, 25, 30, 100]. The mean without the outlier is (10+15+20+25+30+)/5 = 20. However, if we include the 
            outlier (100), the mean becomes (10+15+20+25+30+100)/6 = 33.33. The presence of the outlier significantly increases the mean.
            
    Measures of Dispersion (Range, Standard Deviation):The range, which is the difference between the maximum and minimum values, can be greatly influenced
         by outliers. If there is an extreme value, the range will be larger, indicating greater variability. Similar to the range, the standard deviation 
         is sensitive to outliers. Extreme values can increase the spread of the data, resulting in a larger standard deviation.
            Example: Consider the dataset [5, 10, 15, 20, 25, 100]. The range without the outlier is 100-5 = 95, but with the outlier, it becomes 100-5 = 95
            The standard deviation is also influenced by the outlier, leading to a larger value.

            
           
```
