```python
Q1-Explain the following with an exampl
    (a)Artificial Intelligence
    (b)Machine learning
    (c)Deep Learning
Ans.-(a)AI-It is a smart application that can perform its own task without any human itervention. Example-Self driven car, Robot.
     (b)ML- It is a branch of AI which provide statstool to analyse, visualize and predict model. Example-Recomendatin in amazon and netflix.
     (c)DL- it is subset of ML which is use to mimic human brain.Example-human brain.

Q2- What is supervised learning ? List some examples of supervised learning.
Ans.-Supervised learning-Supervised learning is a type of machine learning in which the model is trained on a labeled dataset, where the input 
     data is associated with a corresponding output variable. The goal of supervised learning is to learn the mapping function that can accurately 
      predict the output variable for new, unseen input data.
    Example.. 
        1.Classification: A model is trained to predict the class or category of a new observation. Examples include image classification, spam 
          detection, and sentiment analysis.
        2.Regression: A model is trained to predict a continuous numerical value. Examples include stock price prediction, house price prediction, 
          and weather forecasting.
        3.Object detection: A model is trained to detect the presence and location of objects within an image or video. Examples include facial 
          recognition, self-driving cars, and security cameras.
        4.Natural language processing (NLP): A model is trained to understand and generate human language. Examples include language translation,
          chatbots, and speech recognition.
        5.Time series prediction: A model is trained to predict future values in a time series. Examples include stock price prediction, energy 
          demand forecasting, and traffic flow prediction.
            
Q3- What is unsupervised learning? List some example of unsupervised learning.
Ans.-Unsupervised learning is a type of machine learning where the model is trained on an unlabeled dataset, without any specific target variable 
     to predict. The goal of unsupervised learning is to identify patterns or relationships in the data, and to group similar data points together.

    Examples of unsupervised learning ...
         1. Clustering: A model is trained to group similar data points together into clusters, based on their features. Examples include customer 
          segmentation, image segmentation, and anomaly detection.
        2.Dimensionality reduction: A model is trained to identify the most important features of a dataset, and to reduce the dimensionality of the
          data while preserving the original information. Examples include principal component analysis (PCA), t-distributed stochastic neighbor 
            embedding (t-SNE), and autoencoders.
        3.Association rule mining: A model is trained to identify relationships between different items or features in a dataset. Examples include 
          market basket analysis, recommendation systems, and fraud detection.
        4.Anomaly detection: A model is trained to identify data points that are significantly different from the majority of the dataset. Examples 
         include fraud detection, intrusion detection, and predictive maintenance.
        5.Generative models: A model is trained to generate new data that is similar to the original dataset. Examples include generative adversarial networks (GANs),
         variational autoencoders (VAEs), and restricted Boltzmann machines (RBMs).
            
            
Q4- What is the difference between AI, ML, DL, and DS?
Ans.-Difference between AI, ML, DL and DS are....
        AI-It is a smart application that can perform its own task without any human itervention. Example-Self driven car, Robot.
        ML- It is a branch of AI which provide statstool to analyse, visualize and predict model. Example-Recomendatin in amazon and netflix.
        DL- it is subset of ML which is use to mimic human brain.Example-human brain.
        DS-DS stands for data science, which is a multidisciplinary field that involves extracting insights and knowledge from data using techniques from 
          statistics, mathematics, and computer science. DS includes tasks such as data cleaning, data wrangling, exploratory data analysis, and predictive
            modeling using techniques from ML and statistical inference.

        in summary , AI is a broad field that encompasses the development of intelligent computer systems, while ML and DL are subfields of AI that focus on 
        developing algorithms that enable machines to learn from data. DS is a field that involves using techniques from statistics and computer science to 
        extract insights from data.
        
        
Q5- What are the main difference between supervised, unsupervised, and semi-supervised learning?
Ans.-Supervised learning-Supervised learning is a type of machine learning in which the model is trained on a labeled dataset, where the input 
     data is associated with a corresponding output variable. The goal of supervised learning is to learn the mapping function that can accurately 
      predict the output variable for new, unseen input data.
        
    Unsupervised learning is a type of machine learning where the model is trained on an unlabeled dataset, without any specific target variable 
     to predict. The goal of unsupervised learning is to identify patterns or relationships in the data, and to group similar data points together.

    Semi-supervised machine learning is a combination of supervised and unsupervised learning. It uses a small amount of labeled data and a large
    amount of unlabeled data, which provides the benefits of both unsupervised and supervised learning while avoiding the challenges of finding a 
    large amount of labeled data.
    
Q6- What is train, test and validation split? Explain the importance of each team.
Ans.-Train Data-Training data is the set of the data on which the actual training takes place. 
    Validation Data-Validation split helps to improve the model performance by fine-tuning the model after each epoch.
    Test Data-The test set informs us about the final accuracy of the model after completing the training phase.
    
Q7- How can unsupervised learning be used in anomaly detection?
Ans.-Unsupervised learning can be used in anomaly detection to identify unusual patterns or outliers in a dataset without requiring labeled data. 
     In unsupervised anomaly detection, the algorithm is trained on the entire dataset, without any prior knowledge of which data points are anomalous
    or normal.
    
     Unsupervised Anomaly Detection...
        1.Load the dataset. 
        2.Check available models. 
        3.Plot model. 
        4.Save the model. 
        5.Load the model. 
        6.Score on unseen data.
        
        
Q8- List down some commonly used supervised algorithms and unsupervised learning
    algorithm?
Ans.-
     Commonly used supervised learning algorithms:
         Regression-1.Linear Regression.
                    2.Polynomial Regression.
                    3.SVR
                    4.Decision Tree
                    5.Random Forest
                    6.Xgboost
                    7.KNN
        Classification-1.Logistic Regression
                    2.SVM
                    3.Decision Tree
                    4.Random Forest
                    5.Native Bayes
                    6.KNN
            
    Commonly used unsupervised learning algorithms:
            1.K-means clustering
            2.Hierarchical clustering
            3.Principal component analysis (PCA)
            4.Anomaly detection
            5.Association rule learning
            6.Self-organizing maps (SOM)
            7.Gaussian mixture models (GMM)
```
