```python
Q1. What is the Probability density function?
Ans. The Probability Density Function (PDF) is a concept in probability theory and statistics that describes the likelihood of a continuous
     random variable taking on a specific value. Probability density function is denoted as f(x)
    
    
Q2. What are the types of Probability distribution?
Ans. The types of Probability distribution are...
    1.Discrete Probability Distributions:
        (1)Bernoulli Distribution
        (2)Binomial Distribution
        (3)Poisson Distribution
        
    2.Continuous Probability Distributions:
        (1)Uniform Distribution
        (2)Normal Distribution (Gaussian Distribution)
        (3)Exponential Distribution
        
    3.Multivariate Distributions:
        (1)Multinomial Distribution
        (2)Multivariate Normal Distribution
        

Q3. Write a Python function to calculate the probability density function of a normal distribution with
    given mean and standard deviation at a given point.
Ans. 
        mean = 0
        std_dev = 1
        x = 0
        pdf_value = normal_pdf(x, mean, std_dev)
        print(f"The value of the PDF at x = {x} for N({mean}, {std_dev}^2) is {pdf_value}")
        
        
Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
    distribution can be applied.
Ans.The binomial distribution is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli
     trials, where each trial has only two possible outcomes: success or failure. The properties of the binomial distribution include..
        1.Fixed Number of Trials (n): The distribution is defined for a fixed number of trials, denoted as n.
        2.Independent Trials: Each trial is assumed to be independent of the others.
        3.Two Possible Outcomes: Each trial has only two possible outcomes, often labeled as success (S) and failure (F).
        4.Constant Probability of Success (p): The probability of success (p) remains constant across all trials.
        5.Discrete Random Variable: The binomial random variable represents the number of successes in n trials, and it takes on integer values.

    Examples of Events where Binomial Distribution Can be Applied:

    1.Coin Flipping:
            Scenario: Consider flipping a fair coin multiple times.
            Outcome: The success might be defined as getting heads on a single flip.
            Parameters: n trials (flips), p=0.5 (probability of getting heads on a single flip).
 
    2.Quality Control:
            Scenario: In a manufacturing process, products are inspected for defects.
            Outcome: The success might be defined as a product passing inspection.
            Parameters: n products inspected, p is the probability of passing inspection, and 1−p is the probability of having a defect.
            
            
            
Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4
    and plot a histogram of the results using matplotlib.
Ans.
    import numpy as np
    import matplotlib.pyplot as plt

    # Parameters for the binomial distribution
    n = 1       # number of trials
    p = 0.4     # probability of success
    size = 1000 # sample size

    # Generate a random sample
    sample = np.random.binomial(n, p, size)

    # Plotting the histogram
    plt.hist(sample, bins=[-0.5, 0.5, 1.5], edgecolor='black', align='mid')
    plt.xticks([0, 1])
    plt.title('Histogram of Binomial Distribution Sample')
    plt.xlabel('Number of Successes')
    plt.ylabel('Frequency')
    plt.show

    
Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
    with given mean at a given point.
Ans. import math

     def poisson_cdf(x, mean):
        """
        Calculate the cumulative distribution function of a Poisson distribution.

        :x: The point at which to evaluate the CDF
        :mean: The mean parameter of the Poisson distribution
        :return: The value of the CDF at x
         """
    
        cdf_value = 0
        for k in range(0, x + 1):
            cdf_value += (math.exp(-mean) * mean**k) / math.factorial(k)
        return cdf_value

    # Example usage
    mean_parameter = 3
    x_value = 2
    cdf_result = poisson_cdf(x_value, mean_parameter)
    print(f"The value of the CDF at x = {x_value} for Poisson({mean_parameter}) is {cdf_result}")
    
    
Q7. How Binomial distribution different from Poisson distribution?
Ans. The Binomial distribution and the Poisson distribution are both probability distributions used to model the number of events, often in the 
     context of counting the number of successes in a series of independent trials. However, they have different characteristics and are applicable 
     in different scenarios.
   
    1.Number of Trials:
        Binomial Distribution: The number of trials (n) is fixed and known in advance.
        Poisson Distribution: The number of trials is not fixed; instead, the distribution is often used to model events occurring in a fixed interval 
                              of time or space.
            
    2.Nature of Outcomes:
        Binomial Distribution: Each trial has two possible outcomes (success or failure).
        Poisson Distribution: It is often used to model rare events where the probability of success is very small.
        
    3.Probability of Success:
        Binomial Distribution: The probability of success (p) is constant across all trials.
        Poisson Distribution: It does not explicitly involve a probability of success; instead, it is characterized by the average rate of event 
        occurrences (λ).

    4.Discreteness:
        Binomial Distribution: The random variable is discrete and represents the count of successes in a fixed number of trials.
        Poisson Distribution: The random variable is also discrete, representing the number of events in a fixed interval, but the interval can be very
        small.

    5.Conditions for Application:
        Binomial Distribution: Appropriate when there are a fixed number of independent trials with two possible outcomes.
        Poisson Distribution: Appropriate when events are rare, the average rate of occurrences is known, and the events are independent.



Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
    sample mean and variance.
Ans.
    import numpy as np

    # Set the parameters
    mean_parameter = 5
    sample_size = 1000

    # Generate a random sample from a Poisson distribution
    poisson_sample = np.random.poisson(mean_parameter, sample_size)

    # Calculate sample mean and variance
    sample_mean = np.mean(poisson_sample)
    sample_variance = np.var(poisson_sample)

    # Print the results
    print(f"Sample Mean: {sample_mean}")
    print(f"Sample Variance: {sample_variance}")


Q9. How mean and variance are related in Binomial distribution and Poisson distribution?
Ans. 
     The mean (μ) and variance (σ**2  ) in probability distributions provide insights into the central tendency and spread of the distribution, 
     respectively, mean and variance are related in the Binomial and Poisson distributions...
        
    1.Binomial Distribution:For a Binomial distribution with parameters n (number of trials) and p (probability of success in each trial):
        Mean (μ):  μ = np
        variance(σ**2): σ**2 = np(1-p)
        
    2.Poisson Distribution:For a Poisson distribution with parameter λ (average rate of events).
        Mean (μ): μ = λ
        Variance (σ**2): σ**2 =  λ
        
        For the Binomial distribution, the variance is influenced by both the number of trials (n) and the probability of success (p). When 
    p is close to 0 or 1, the variance is relatively small.
        For the Poisson distribution, the variance is solely determined by the average rate of events (λ). The variance is equal to the mean, 
    which suggests that the spread of the distribution is directly linked to the average rate of event occurrences.
    
    
    
Q10. In normal distribution with respect to mean position, where does the least frequent data appear?
Ans. In a normal distribution, also known as a Gaussian distribution or bell curve, the data is symmetrically distributed around the mean. The mean 
    represents the central point of the distribution, and the shape of the curve is such that data points are more concentrated around the mean and 
    become less frequent as you move away from the mean in either direction.
```
