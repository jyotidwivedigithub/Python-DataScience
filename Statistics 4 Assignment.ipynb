```python
Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
    an example.
Ans. The Probability Mass Function (PMF) and Probability Density Function (PDF) are both concepts in probability theory that describe the likelihood 
    of a random variable taking on a specific value or falling within a certain range.    
    
    1.Probability Mass Function (PMF):The PMF is used to describe the probability distribution of a discrete random variable. It gives the probability 
      of each possible value that the discrete variable can take. Mathematically, the PMF is denoted asP(X=x), where X is the random variable and x is
        a specific value it can take.
        Example: Rolling a dice
    
    2.Probability Density Function (PDF):The PDF is used to describe the probability distribution of a continuous random variable. Unlike the PMF, 
      the PDF does not give the probability of specific values but rather the probability density at a given point. Mathematically, the PDF is denoted
        as f(x), and the probability of the variable falling within a specific range [a,b] is given by the integral of the PDF over that range:
        Example: Standard Normal Distribution.
        
        
        
Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?
Ans.The Cumulative Distribution Function (CDF) is a concept in probability theory that provides the probability that a random variable takes on a 
    value less than or equal to a given point. It gives the cumulative probability up to a certain value of the variable. Mathematically, the CDF is 
    denoted as F(x).
          F(x) = p(X<=x)
     
    Example:Let's consider an example using a six-sided fair die. The CDF for rolling a fair die, denoted as F(x), where x is a specific value, can be
            calculated as follows:
            F(1)=P(X≤1)= 1/6(probability of rolling a 1 or less)
            F(1)=P(X≤2)= 2/6= 1/3(probability of rolling a 2 or less)
            F(1)=P(X≤3)= 3/6= 1/2(probability of rolling a 3 or less)
            F(1)=P(X≤4)= 4/6= 2/3(probability of rolling a 4 or less)
            F(1)=P(X≤5)= 5/6(probability of rolling a 5 or less)
            F(1)=P(X≤6)= 6/6=1(probability of rolling a 6 or less)
            
    CDF is used to provide...
        (1) Cummulative Informtion
        (2) Quantifying Probability
        (3) Statistical Analysis
        (4) Visualization
            
            
Q3: What are some examples of situations where the normal distribution might be used as a model?
    Explain how the parameters of the normal distribution relate to the shape of the distribution.
Ans. 
    The normal distribution, also known as the Gaussian distribution, is a widely used probability distribution in statistics. It is characterized by 
    its bell-shaped curve, which is symmetrical around the mean. The parameters of the normal distribution are the mean (μ) and the standard deviation
    (σ). Some examples of situations where the normal distribution might be used as a model:

    1.Height of a Population:
        Example: The heights of adult males in a given population.
        Parameters: Mean height (μ) and standard deviation (σ) would describe the average height and the variation around that average.
    2.IQ Scores:
        Example: Intelligence Quotient (IQ) scores in a population.
        Parameters: Mean IQ (μ) and standard deviation (σ) would indicate the average intelligence level and the spread of IQ scores in the population.
    3.Measurement Errors:
        Example: Errors in measurements due to various factors.
        Parameters: Mean error (μ) and standard deviation (σ) would describe the average measurement error and the variability of errors.
    4.Test Scores:
        Example: Scores on a standardized test.
        Parameters: Mean score (μ) and standard deviation (σ) would represent the average performance and the spread of scores.
    5.Financial Returns:
        Example: Daily returns of a stock or financial instrument.
        Parameters: Mean return (μ) and standard deviation (σ) would describe the average daily return and the volatility of the financial instrument.
    6.Natural Phenomena:
        Example: Measurement errors in natural processes.
        Parameters: Mean measurement (μ) and standard deviation (σ) would describe the average measurement and the variability in natural phenomena.

        The shape of the normal distribution is determined by its mean and standard deviation:
            1.The mean (μ) is the central point of the distribution, and it represents the location of the peak of the curve.
            2.The standard deviation (σ) controls the spread or width of the distribution. A larger standard deviation results in a wider and flatter
             curve, while a smaller standard deviation leads to a narrower and taller curve.
                
                
Q4: Explain the importance of Normal Distribution. Give a few real-life examples of Normal
    Distribution.
Ans. The normal distribution is important in statistics and data analysis for several reasons:
        1.Widespread Applicability: The normal distribution is a versatile model that is applicable to a wide range of real-world phenomena. Many
          natural and human-made processes exhibit a distribution of values that approximates a normal distribution.
        2.Central Limit Theorem: The normal distribution is a key component of the Central Limit Theorem. According to this theorem, the distribution
          of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless
          of the original distribution of the variables. This makes the normal distribution a fundamental concept in inferential statistics.
        3.Parameter Estimation: In many statistical methods, the normal distribution is used as an assumption for parameter estimation. Maximum Likelihood
          Estimation (MLE) and other estimation techniques often assume a normal distribution for the underlying data.  
        4.Statistical Testing: Many statistical tests and procedures are based on the assumption of normality. For example, t-tests, analysis of 
          variance (ANOVA), and linear regression often assume normality in the residuals for valid interpretations of results.
        5.Prediction and Control: In fields such as quality control and manufacturing, the normal distribution is used to model the variability of 
          measurements. This allows for the prediction of expected outcomes and the setting of control limits to identify unusual observations.
            
    Real-life examples of normal distribution include:
        1.Human Characteristics
        2.Financial Markets
        3.Natural Processes:
        4.Quality Control:


Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
    Distribution and Binomial Distribution?
Ans. The Bernoulli distribution is a discrete probability distribution that models a random experiment with two possible outcomes, typically labeled 
    as success (coded as 1) and failure (coded as 0). It is named after Jacob Bernoulli, a Swiss mathematician. The distribution is characterized by a
    single parameter, denoted as p, which represents the probability of success.
    Example:
        Consider a single toss of a fair coin, where we define getting heads as a success (coded as 1) and getting tails as a failure (coded as 0).
        The outcome of this experiment follows a Bernoulli distribution with p=0.5, as the probability of getting heads is 0.5.
    
    Bernoulli vs. Binomial Distribution:
        1.Bernoulli Distribution: 
            *Models a single trial or experiment with two possible outcomes (success or failure).
            *Parameters: p, the probability of success.
            *Notation: X∼Bernoulli(p).
            
        2.Binomial Distribution:
            *Models the number of successes in a fixed number of independent and identical Bernoulli trials.
            *Parameters: n, the number of trials, and p, the probability of success in each trial.
            *Notation: X∼Binomial(n,p). 
            
            
Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
    is normally distributed, what is the probability that a randomly selected observation will be greater
    than 60? Use the appropriate formula and show your calculations.
Ans. 
    To find the probability that a randomly selected observation from a normally distributed dataset is greater than 60, given that the mean (μ) 
    is 50 and the standard deviation (σ) is 10, we can use the properties of the normal distribution.

    First, we standardize the value 60 using the Z-score formula:
        Z= (X−μ)/ σ
    where:
        (1)X is the value we're interested in (60 in this case),
        (2)μ is the mean (50), and
        (3)σ is the standard deviation (10).
        Z = (60-50)/10 = 1
    So, the Z-score of 60 is 1. This means that 60 is one standard deviation above the mean in this distribution.
    
    The Z-score tells us how many standard deviations away from the mean our value is. To find the probability that a randomly selected observation is
    greater than 60, we need to find the area to the right of Z = 1 in the standard normal distribution.

    This is done using a standard normal distribution table or a statistical software. Standard normal distribution tables give the area to the left of
    a Z-score. For Z = 1, this area is approximately 0.8413. However, we want the area to the right (since we're interested in values greater than 60).

       The area to the right is given by:
            P(X>60)=1−P(X≤60)
            P(X>60)=1−0.8413
            P(X>60)=0.1587

    So, the probability that a randomly selected observation from this dataset is greater than 60 is approximately 0.1587, or 15.87%.                                                                                              
                                                     
                                                                                                     
                                                                                                     
Q7: Explain uniform Distribution with an example.
Ans. The uniform distribution is a probability distribution in which all outcomes are equally likely. It is characterized by a constant probability for
    all possible values within a specified range. The probability density function (PDF) of a continuous uniform distribution is flat, indicating that 
    any value within the range has an equal chance of occurring.
    The probability density function for a continuous uniform distribution between a and b is given by:
          f(x) = 1/(b-a)
    where:
      * a is the lower bound of the distribution,
      * b is the upper bound of the distribution,
      * f(x) is the probability density function.  
                
    Example:
       Let's consider an example of a continuous uniform distribution representing the rolling of a fair six-sided die. In this case, the possible 
       outcomes are integers from 1 to 6, and each outcome has an equal probability of 1/6. The range of the distribution is from 1 to 6.  
       (1) Lower bound(a):1
       (2) Upper bound(b):6      
                                                                                                     
    The probability density function for this uniform distribution is: f(x)= 1/6.
    This means that any specific outcome (1, 2, 3, 4, 5, or 6) has a probability of 1/6 of occurring when rolling the die.
                                                                                                     
                                                                                                                                                                                                                                                                                                             
Q8: What is the z score? State the importance of the z score.
Ans.The Z-score, also known as a standard score, is a statistical measurement that describes a value's relationship to the mean of a group of values, 
    measured in terms of standard deviations from the mean. The Z-score is calculated using the following formula:
         Z= (X−μ)/σ
   where:
      * X is the value being measured,
      * μ is the mean of the population or sample,
      * σ is the standard deviation of the population or sample.
    
  Importance of the Z-score:
      * Standardization
      * Identification of outliers
      * Normal Distribution Properties                                                                                              
      * Comparison Across Different Populations
      * Use in Statistical Tests                                                                                               
              
                                                                                                     
                                                                                                     
Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.
Ans. The Central Limit Theorem (CLT) is a fundamental concept in statistics that describes the shape of the sampling distribution of the sample mean
    for a sufficiently large sample size, regardless of the shape of the population distribution. In other words, as the sample size increases, the 
    distribution of sample means approaches a normal distribution, even if the original population distribution is not normal.   
       
      The Central Limit Theorem is stated as follows:                                                                                               
        1.For a Random Sample.
        2.Large Sample size.
        3.The Central Limit Theorem is often stated as follows.
                                                                                                     
      The significance of the Central Limit Theorem lies in...
        1.Statistical Inference.
        2.Sampling Distribution of the mean.
        3.Confidence intervals.                                                                                             
        4.Hypothesis Testing.
        5.Practical Applications.                                                                                             
        6.Simulating Normality.                                                                                            
                          
                                                                                                     
                                                                                                     
Q10: State the assumptions of the Central Limit Theorem.
Ans. The main assumptions of the Central Limit Theorem are...
       1.Independence of observations.                                                                                              
       2.Sample size or Number of trials.
       3.Population's distribution.
       4.Random sampling.                                                                                              
                                                                                                     
```
