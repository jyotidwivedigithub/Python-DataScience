```python
Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.
And.
    In Bokeh, glyphs are graphical shapes that can be added to a plot to represent data. Glyphs can be used to represent 
    various types of data, including points, lines, bars, and patches. Glyphs can be customized with different colors, sizes,
    and styles to make the plot more visually appealing and informative.
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
colors = ["red", "green", "blue", "orange", "purple"]
sizes = [10, 20, 30, 40, 50]

output_file("bokeh_glyphs.html")
p = figure(title="Bokeh Glyphs", x_axis_label='X-axis', y_axis_label='Y-axis')

# Create a ColumnDataSource object
source = ColumnDataSource(data={'x': x, 'y': y, 'colors': colors, 'sizes': sizes})

# Add circle glyphs to the figure
p.circle(x='x', y='y', color='colors', size='sizes', source=source)
show(p)
```

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, output_file, show
from bokeh.models import Legend, LegendItem

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
colors = ["red", "green", "blue", "orange", "purple"]
sizes = [10, 20, 30, 40, 50]
labels = ["Label 1", "Label 2", "Label 3", "Label 4", "Label 5"]

output_file("bokeh_customization.html")

fig = figure(title="Bokeh Customization", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add circle glyphs to the figure
fig.circle(x, y, color=colors, size=sizes, legend_label="Circles")

# Customize the appearance of the plot
p.title.text_color = "blue"
p.title.text_font_size = "20pt"
p.xaxis.axis_label_text_color = "green"
p.yaxis.axis_label_text_color = "green"
p.xaxis.major_label_text_color = "red"
p.yaxis.major_label_text_color = "red"

# Add a legend to the plot
legend = Legend(items=[LegendItem(label="Circles", renderers=[fig.renderers[0]])])
p.add_layout(legend, 'right')
show(p)
```

```python
Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated inreal time?
Ans.
    Bokeh is a Python library that allows for interactive visualization of data in web browsers. Bokeh server is a 
    component of Bokeh that allows for creating and deploying interactive web applications that can update in real time. 
      To create an interactive plot with Bokeh server, you need to define a Bokeh application that includes the plot and 
    any widgets you want to use. You can define the layout of the application using the Bokeh layout functions, such as column,
    row, and grid.
      Once you have defined your application, you can start the Bokeh server and deploy your application using the bokeh serve 
    command. This will launch a web server that serves your application, and users can access it using a web browser.
      As users interact with the widgets in your application, the Bokeh server will update the plot in real time, allowing users 
    to explore the data and see the effects of their changes immediately.
      Bokeh server is a powerful tool used for creating interactive web applications that allow users to explore data in real time. 
    By defining a Bokeh application that includes interactive plots and widgets, you can create a dynamic and engaging user experience 
    that encourages exploration and discovery.
    
```

```python
Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?
```

```python
#Embedding Bokeh plots in Flask
#Flask is a popular Python web framework that allows to create web applications easily. 
To embed Bokeh plots in Flask, you can use the bokeh.embed module.
from flask import Flask, render_template
from bokeh.plotting import figure
from bokeh.embed import components

app = Flask(__name__)

@app.route('/')
def index():
    # Create a Bokeh plot
    plot = figure()
    plot.circle([1, 2, 3], [4, 5, 6])

    # Embed the plot into the HTML template
    script, div = components(plot)
    return render_template('index.html', script=script, div=div)

if __name__ == '__main__':
    app.run()
```

```python
#Embedding Bokeh plots in Django
#Django is another popular Python web framework that can be used to create web applications. To embed Bokeh plots 
# in Django, you can use the bokeh.resources and bokeh.embed modules.
from django.shortcuts import render
from bokeh.plotting import figure
from bokeh.resources import CDN
from bokeh.embed import components

def index(request):
    # Create a Bokeh plot
    plot = figure()
    plot.circle([1, 2, 3], [4, 5, 6])

    # Embed the plot into the HTML template
    script, div = components(plot, CDN)

    return render(request, 'index.html', {'script': script, 'div': div})
```
