# Standardization

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bills = list(df['total_bill'])
```

```python
total_bills
```

```python
sns.histplot(total_bills)
```

```python
import numpy as np
mean = np.mean(total_bills)
std = np.std(total_bills)
mean, std
```

```python
normalized_data = []
for i in total_bills :
    z_score = (i-mean)/std   
    normalized_data.append(z_score)
    
print(normalized_data)
```

```python
sns.histplot(normalized_data)
```

# Normalization - Min Max Scaler (0-1)

```python
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max.fit(df[['distance','fare','tip']])
```

```python
min_max.transform(df[['distance','fare','tip']])
```

```python
min_max.fit_transform(df[['distance','fare','tip']])
```

```python
min_max.transform([[1,3,4]])
```
