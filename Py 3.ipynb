```python
def word_count(text):
    # Create an empty dictionary
    word_dict = {}
    # Split the input string into words
    words = text.split()
    # Loop over the words and count their occurrences
    for word in words:
        if word in word_dict:
            word_dict[word] += 1
        else:
            word_dict[word] = 1
    # Return the dictionary
    return word_dict

# Example usage
text = "hello world hello"
counts = word_count(text)
print(counts)
```
