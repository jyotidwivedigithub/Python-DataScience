```python
# Merge...work on columns
# left, right, outer,cross, inner
pd.merge(data1,data2)      # intersection-inner merge
```

```python
pd.merge(data1,data2,how = 'left')
```

```python
pd.merge(data1,data2,how = 'right')
```

```python
pd.merge(data1,data2,how = 'outer')
```

```python
pd.merge(data1,data2,how = 'cross')
```

```python
# Apply function is used when we have money in Rs we want to
# convert in to any other currency like INR, EURO, DOLLAR

df['HourlyRate_INR'] = df['HourlyRate'].apply(lambda x : x*80)
```

```python
# OR
def euro_inr(x):
    return x*80

df['HourlyRate_INR'] = df['HourlyRate'].apply(euro_inr)
```

```python
def Job_Satisfaction(x):
    if x<3:
        return "Good"
    elif x>+3 and x<5:
        return "VeryGood"
    else:
        return "Nice"
        
```

```python
df['Job_Satisfaction'] = df['JobSatisfaction'].apply(Job_Satisfaction)
```
