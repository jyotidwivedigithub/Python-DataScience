```python
# SCATTER PLOT
x = np.random.rand(50)
y = np.random.rand(50)
```

```python
colours = np.random.rand(50)     # to show colourfull data
sizes = 1000*np.random.rand(50)  # for different data-sizes 100 is size of data
plt.xlabel("x-axis")             # alpha represent intensity of colour
plt.ylabel("y-axis")
plt.title("random data")
plt.scatter(x,y , c = colours , s=sizes, alpha=.5)
```

```python
# BAR GRAPH

x = ['a','b','c','d','e']   # 5-categorical (string)
y = np.random.rand(5)       # 5-numerical value
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.bar(x,y)                # vertical graph
```

```python
y
```

```python
plt.barh(x,y)          #barh for horizontal graph
```

```python
x = np.linspace(0,10,200)
y = np.sin(x)

plt.figure(figsize=(6,2))
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.plot(x,y, "--b")        # dots plot graph design .,--,*,etc
plt.grid()                  # for background box design
plt.show()
```

```python
# SCATTER PLOT
x = np.random.rand(50)
y = np.random.rand(50)

colors = np.random.rand(50)     # to show colourfull data
sizes = 1000*np.random.rand(50)  # for different data-sizes 100 is size of data
plt.xlabel("x-axis")             # alpha represent intensity of colour
plt.ylabel("y-axis")
plt.title("random data")         # marker= 'v','x','*','o'
plt.scatter(x,y ,c = colors ,s=sizes ,alpha=.5, marker='v')
```

```python
# DATA PROJECTION
# Data Projection is possible on 3-d not on 2-d

x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

colors = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z, c=colors )
ax.show()
```

```python
# BOX PLOT
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

fig, ax = plt.subplots()
ax.boxplot(data)

ax.set_xticklabels(['Box 1', 'Box 2'])
ax.set_ylabel('Value')

plt.show()
```
