# Decision Tree Classifier

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
dataset = load_iris()
```

```python
print(dataset.DESCR)
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
dataset.target
```

```python
# Independent and dependent features
X=df.iloc[:,:-1]
y=dataset.target
```

```python
X,y
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

## Post Pruning (Small_dataset)

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifier = DecisionTreeClassifier()
treeclassifier.fit(X_train, y_train)
```

```python
X_train.head()
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled=True)
```

```python
# Again Post Pruning with max_depth=2

treeclassifier = DecisionTreeClassifier(max_depth=2)
treeclassifier.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier,filled=True)
```

```python
# Prediction
y_pred = treeclassifier.predict(X_test)
y_pred
```

```python
# Accuracy_Sore and Classifier_report

from sklearn.metrics import accuracy_score, classification_report
score = accuracy_score(y_pred, y_test)
print(score)
print(classification_report(y_pred, y_test))
```

```python
classification_report(y_pred, y_test)
```

## Decision Tree Pre Pruning (for Large_dataset) and Hyperparameter Tuning 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Finding which parameter is suitable
parameter={
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'splitter'  : ['best', 'random'],
    'max_depth' : [1,2,4,5],
    'max_features': ['auto', 'sqrt', 'log2']
}
    
```

```python
## Model_selection

from sklearn.model_selection import GridSearchCV        # CV = Cross_Validation
treeclassifier = DecisionTreeClassifier()
clf = GridSearchCV(treeclassifier, param_grid=parameter, cv=5, scoring='accuracy')
```

```python
# Train the dataset
clf.fit(X_train, y_train)
```

```python
clf.best_params_
```

```python
# Prediction
y_pred = clf.predict(X_test)
y_pred
```

```python
# Accuracy_score
score = accuracy_score(y_pred, y_test)
print(score)
```

```python
# Classification Report
print(classification_report(y_pred, y_test))
```
