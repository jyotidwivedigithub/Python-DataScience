```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create synthetic data points
from sklearn.datasets import make_classification

X,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2,
                          n_redundant=0)
```

```python
X
```

```python
y
```

```python
pd.DataFrame(X)[0]
```

```python
sns.scatterplot(x=pd.DataFrame(X)[0], y=pd.DataFrame(X)[1], hue=1)
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.25, random_state=10)
```

## SVC-Support Vector Classifier

```python
from sklearn.svm import SVC
svc = SVC(kernel='linear')
svc.fit(X_train,y_train)
```

```python
# Coefficient
svc.coef_
```

```python
# Prediction
y_pred = svc.predict(X_test)
y_pred
```

```python
# Classification_report, Confusion_matrix, Accuracy_score

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```

```python
# Hyperparameter Tuning with SVC

from sklearn.model_selection import GridSearchCV         # CV-Cross validation

# defining parameter range
param_grid = { 'C' : [0.1, 1, 10, 100, 1000],            # C-Classifier
              'gamma' : [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel' : ['linear']
             }
```

```python
grid = GridSearchCV(SVC(), param_grid=param_grid, refit=True, cv=5, verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
# Prediction
y_pred1 = grid.predict(X_test)
y_pred1
```

```python
# Classification report, confusion matrix, accuracy score
print(classification_report(y_test, y_pred1))
print(confusion_matrix(y_test, y_pred1))
print(accuracy_score(y_test, y_pred1))
```
